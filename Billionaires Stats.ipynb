```python
import pandas as pd
df = pd.read_csv('Billionaires Statistics Dataset.csv', encoding='latin1')
df.head(5)
```

```python
df.isnull().sum()
```

```python
columns_to_drop = ['organization', 'title', 'state','residenceStateRegion']
df = df.drop(columns_to_drop, axis=1)
```

```python
df = df.dropna(axis=0)
```

```python
import matplotlib.pyplot as plt
category_counts = df['category'].value_counts()
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Individuals by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['age'])
plt.ylabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
country_counts = df['country'].value_counts()[:15]
plt.bar(country_counts.index, country_counts.values)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Distribution of Individuals by Country (Top 15)')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['age'], df['finalWorth'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Final Worth')
plt.title('Scatter Plot of Age vs. Final Worth')
plt.show()
```

```python
import matplotlib.pyplot as plt

category_selfMade_counts = df.groupby(['category', 'selfMade']).size().unstack()
category_selfMade_counts.plot(kind='bar', stacked=True)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Self-Made Status within Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

industry_counts = df['industries'].value_counts()[:15]
plt.bar(industry_counts.index, industry_counts.values)
plt.xlabel('Industry')
plt.ylabel('Count')
plt.title('Distribution of Individuals by Industry (Top 15)')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

category_gender_counts = df.groupby(['category', 'gender']).size().unstack()
category_gender_counts.plot(kind='bar', stacked=True)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Gender within Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='category', y='age')
plt.xlabel('Category')
plt.ylabel('Age')
plt.title('Age Distribution by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df[['finalWorth', 'age', 'gdp_country', 'cpi_country']], diag_kind='kde')
plt.suptitle('Pairwise Relationships')
plt.show()
```
